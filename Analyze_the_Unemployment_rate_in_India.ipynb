# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

# Load the Dataset

```python
url = 'https://github.com/tejas-52/Oasis-Data-Science-Internship/raw/main/Unemployment%20in%20India.csv'
unemployment_data = pd.read_csv(url)
```

# Explore the Dataset

```python
# Display the first few rows of the dataset
print(unemployment_data.head())

# Get a summary of the dataset
print(unemployment_data.info())

# Check for missing values
print(unemployment_data.isnull().sum())

# Get basic statistics of the dataset
print(unemployment_data.describe())
```

# Data Preprocessing

```python
# Display the column names to check for any issues
print(unemployment_data.columns)

# Strip any leading or trailing spaces from the column names
unemployment_data.columns = unemployment_data.columns.str.strip()

# Strip leading and trailing spaces from the 'Date' column values
unemployment_data['Date'] = unemployment_data['Date'].str.strip()

# Convert the 'Date' column to datetime format
unemployment_data['Date'] = pd.to_datetime(unemployment_data['Date'], format='%d-%m-%Y')

# Check the data types to confirm the conversion
print(unemployment_data.dtypes)

# Drop any rows with missing values (if necessary)
unemployment_data.dropna(inplace=True)
```

# Visualize the Data

```python
# Drop non-numeric columns from the dataframe
numeric_data = unemployment_data.select_dtypes(include=np.number)

# Heatmap to visualize correlations between numerical columns
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Analyze Trends and Insights

```python
# Calculate the mean unemployment rate for different regions
region_mean_unemployment = unemployment_data.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
print(region_mean_unemployment)

# Calculate the monthly average unemployment rate
unemployment_data['Month'] = unemployment_data['Date'].dt.to_period('M')
monthly_avg_unemployment = unemployment_data.groupby('Month')['Estimated Unemployment Rate (%)'].mean()
print(monthly_avg_unemployment)

# Plot the monthly average unemployment rate
plt.figure(figsize=(12, 6))
monthly_avg_unemployment.plot(kind='bar')
plt.title('Monthly Average Unemployment Rate')
plt.xlabel('Month')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()
```
